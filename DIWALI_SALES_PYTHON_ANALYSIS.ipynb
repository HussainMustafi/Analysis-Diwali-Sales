# Exploratory Data Analysis

# Gender

```python

df.columns
```

```python
gc=sns.countplot(x='Gender',data=df)
sns.set(rc={'figure.figsize':(6,5)})
#adding value lebels to chart
for bars in gc.containers:
    gc.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_by_Gender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data= sales_by_Gender)
```

From the above two graphs we can see that most of the customer were female and most sales are also generated by female

```python
df.columns
```

# age

```python
age_grp=sns.countplot(x='Age Group',data=df,hue='Gender') 
# we use hue to separate the each age_group by perticular gender group
sns.set(rc={'figure.figsize':(15,5)})
#adding value lebels to chart
for bars in age_grp.containers:
    age_grp.bar_label(bars)
```

```python
#total  amount vs sales group
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data= sales_age)
```

most of the buyers are between the age group of 26-35 and most of them are female


# state

```python
# total number of orders from the  states
sales_states = df.groupby(['State'], as_index =False)['Orders'].sum().sort_values(by='Orders',ascending=False)
```

```python
#top 10 states
sales_states.head(10)
```

```python
sns.barplot( x='State',y='Orders', data= sales_states.head(10))
sns.set(rc={'figure.figsize':(15,5)}) #to adjust the width and height of the graph
```

```python
#total amount/sales from the top 10 states
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
```

```python
sns.barplot(x='State',y='Amount', data = sales_state)
```

from the above two graphs we can see that most of the orders and sales are coming from UP ,Maharashtra and KA .


but unexpectedly instead of having much larger amount of order quantites, Haryana has generated more amount of sales than Kerala

# Marital Status

```python
gc=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(3,5)})
#adding value lebels to chart
for bars in gc.containers:
    gc.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

```

```python
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state, x='Marital_Status',y='Amount',hue="Gender")
 
```

from above two graphs we can tell that the most sales are generated by married women

# Product _category

```python
b = sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(30,10)})
for bars in b.containers:
    b.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
```

```python
sns.barplot(x='Product_Category',y='Amount',data=sales_state)
sns.set(rc={'figure.figsize':(10,5)})
```

from the above two graphs we can see that most sales are generated by Food although Clothing and Apparel has the most number of product sold

```python
#finding the best selling product
sales_state = df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
```

```python
sns.barplot(x='Product_ID',y='Amount',data=sales_state)
sns.set(rc={'figure.figsize':(5,5)})
```
